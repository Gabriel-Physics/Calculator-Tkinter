```python
from tkinter import *
import tkinter.font as font


from pip._vendor.colorama import Fore, init





root=Tk()
root.title ('My Calculator')
root.geometry ('260x500')
#root.iconbitmap('C:/users/gabriel/desktop/python/icone_zoppo2.ico')

fonte=font.Font(size=12,weight='bold')

quadro = LabelFrame (root, text="Testcalc",padx=10,pady=10)
quadro.pack(padx=20,pady=20)

def botao_numero(n,erro):    #insere numero ou ponto decimal na tela. Se deu erro antes, apaga o erro do visor antes de escrever o numero
    global error
    if erro:
        mostrador.delete(0,END)
        error = False
    atual=mostrador.get()
    mostrador.delete (0, END)
    mostrador.insert(0, atual+str(n))

def botao_apagar():   #apaga o mostrador
    mostrador.delete(0, END)

def botao_somar():    
    global prim_num
    global operacao
    operacao = 1
    prim_num = mostrador.get()
    mostrador.delete(0, END)
    

def botao_subtrair():
    global prim_num
    global operacao
    operacao = 2
    prim_num = mostrador.get()
    mostrador.delete(0, END)

def botao_multiplicar():
    global prim_num
    global operacao
    operacao = 3
    prim_num=mostrador.get()
    mostrador.delete(0, END)

def botao_dividir():
    global prim_num
    global operacao
    operacao = 4
    prim_num=mostrador.get()
    mostrador.delete(0, END)



def botao_igual(operacao):
    
    segundo_numero = mostrador.get()

    if operacao == 1:
        soma = float(prim_num) + float (segundo_numero)
        mostrador.delete(0, END)
        mostrador.insert(0, soma)
        # operacao=0
    elif operacao == 2:
        subtracao = float(prim_num) - float (segundo_numero)
        mostrador.delete(0, END)
        mostrador.insert(0, subtracao)
        # operacao=0

    elif operacao == 3:
        multiplicacao = float(prim_num) * float (segundo_numero)
        mostrador.delete(0, END)
        mostrador.insert(0, multiplicacao)
        # operacao=0

    elif operacao == 4:
        try:
            quociente = float(prim_num) / float (segundo_numero)    
        except ZeroDivisionError:
            global error
            error=True
            mostrador.delete(0,END)
            mostrador.insert(0,'Divis√£o por zero!')
            print (Fore.RED + 'askjaslfasdf')
            #time.sleep (10)         
            # operacao=0

        else:
            quociente = float(prim_num) / float (segundo_numero)
            mostrador.delete(0, END)
            mostrador.insert(0, quociente)
            # operacao=0




primeiro_numero = 0
error = False



Button(quadro, text='7', padx=10,pady=10, font=fonte, command = lambda:botao_numero(7,error) ).grid (row=2,column=0)
Button(quadro, text='8', padx=10,pady=10, font=fonte,command = lambda:botao_numero(8,error) ).grid (row=2,column=1)
Button(quadro, text='9', padx=10,pady=10,font=fonte, command = lambda: botao_numero(9,error)).grid (row=2, column=2)

Button(quadro, text='4', padx=10,pady=10,font=fonte,command = lambda:botao_numero(4,error)).grid (row=3,column=0)
Button(quadro, text='5', padx=10,pady=10,font=fonte,command = lambda:botao_numero(5,error)).grid (row=3,column=1)
Button(quadro, text='6', padx=10,pady=10,font=fonte,command = lambda:botao_numero(6,error)).grid (row=3,column=2)

Button(quadro, text='1', padx=10,pady=10,font=fonte,command = lambda:botao_numero(1,error)).grid (row=4,column=0)
Button(quadro, text='2', padx=10,pady=10,font=fonte,command = lambda:botao_numero(2,error)).grid (row=4,column=1)
Button(quadro, text='3', padx=10,pady=10,font=fonte,command = lambda:botao_numero(3,error)).grid (row=4,column=2)

Button(quadro, text='0', padx=10,pady=10,font=fonte,command = lambda:botao_numero(0,error)).grid (row=5,column=0)
Button(quadro, text='.', padx=10,pady=10,font=fonte,command = lambda:botao_numero('.',error)).grid (row=5,column=1)
Button(quadro, text='Erase', padx=3,pady=10, bg='white',command=botao_apagar).grid (row=5,column=2)

Button(quadro, text='-', padx=10,pady=10,bg='grey',font=fonte, command = botao_subtrair).grid (row=6,column=0)
Button(quadro, text='X', padx=10,pady=10,bg='grey', font=fonte,command = botao_multiplicar).grid (row=6,column=1)
Button(quadro, text='/', padx=10,pady=10, bg='grey',font=fonte, command = botao_dividir).grid (row=6,column=2)

Button(quadro, text='+', padx=10,pady=10, bg='grey', font=fonte,command = botao_somar).grid (row=7,column=0)
Button(quadro, text='=', padx=39,pady=10,bg='green' ,font=fonte, command =lambda:botao_igual(operacao)).grid (row=7,column=1,columnspan=2)

Button (root, text='Exit', padx=5,pady=5,bg='blue', font=fonte,command=root.quit).pack()

# Foto
#foto = ImageTk.PhotoImage (Image.open('zoppopic.jpg'))
 

fotorig = fotorig.resize ((280,200))
foto = ImageTk.PhotoImage (fotorig)
    
    
    

# Label (quadro, image=foto).grid(row=0,column=0,columnspan=3)


mostrador = Entry (quadro, width=10, bg='yellow',borderwidth=5, font=("Arial", 18), cursor="dot")
mostrador.grid(row=1, column=0, columnspan=3)




root.mainloop()
```
